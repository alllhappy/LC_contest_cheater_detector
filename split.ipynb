convert into suitable data types and handling null values

```python

for i in range(1,5):
    df1[f'Q{i}FailCnt'] = df1[f'Q{i}FailCnt'].fillna(0).astype('int64')
    df1[f'Q{i}TimeStamp'] = df1[f'Q{i}TimeStamp'].fillna(0).astype('int64')
df1['CurrRating'] = df1['CurrRating'].astype('int64')
df1.head()
```

```python

for i in range(1,5):
    df2[f'Q{i}FailCnt'] = df2[f'Q{i}FailCnt'].fillna(0).astype('int64')
    df2[f'Q{i}TimeStamp'] = df2[f'Q{i}TimeStamp'].fillna(0).astype('int64')
df2['CurrRating'] = df2['CurrRating'].astype('int64')
df2.head()
```

```python
for i in range(1,5):
    df3[f'Q{i}FailCnt'] = df3[f'Q{i}FailCnt'].fillna(0).astype('int64')
    df3[f'Q{i}TimeStamp'] = df3[f'Q{i}TimeStamp'].fillna(0).astype('int64')
df3['CurrRating'] = df2['CurrRating'].astype('int64')
df3.head()
```

time conversion from EPOCH to UTC time

```python

#opetional can directly subtract epoch time to find differences in second and find time for each question
df1['FinishTimeStamp']=pd.to_datetime(df1['FinishTimeStamp'],unit='s')
df2['FinishTimeStamp']=pd.to_datetime(df2['FinishTimeStamp'],unit='s')
df3['FinishTimeStamp']=pd.to_datetime(df3['FinishTimeStamp'],unit='s')

def timeConvert(df):
    for i in range(1,5):
        df[f'Q{i}TimeStamp']=pd.to_datetime(df[f'Q{i}TimeStamp'],unit='s')
    return df


df1=timeConvert(df1)
df1.head()
```

```python
df2=timeConvert(df2)
df3=timeConvert(df3)
```

test train Split
will label this test data for model evaluation
Also Make Ensure that training data is used for training purposes

ensure no data leakage. if a person is in testing data set make sure present in other training (possible when the person has participated in 2 or more contest)

```python
from sklearn.model_selection import train_test_split

# 60 row from each contest should be used
X1_train, X1_test = train_test_split(
    df1,
    test_size=0.0065, #will give 60 rows
    random_state=41
)

X2_train, X2_test = train_test_split(
    df2,
    test_size=0.0063, #will give 60 rows
    random_state=41
)

X3_train, X3_test = train_test_split(
    df3,
    test_size=0.0063, #will give 60 rows
    random_state=41
)

print(X1_test.shape)
print(X2_test.shape)
print(X3_test.shape)

```

```python
X_test=pd.concat([X1_test,X2_test,X3_test],ignore_index=True) #combine all test dataset into 1
X_test.shape
```

see Rank distribution in testing data

```python
X_test.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
ranges=[]
lb=[]
for i in range(1,10000,1000):
    ranges.append(i-1)

def showDist(X_t):
    X=X_t.copy()
    X['range'] = pd.cut(X['Rank'], ranges)
    X['range'].value_counts().sort_index().plot(kind='bar') #histogram has no gaps b/w them also is more suitable for continuous data distribution
    plt.xlabel('Rank Range')
    plt.ylabel('Count')
    plt.title('Distribution of Rank Ranges in Test Dataset')
    plt.xticks(rotation=80)
```

```python
showDist(X_test)
print(f'minium Rank = {X_test["Rank"].min()},maxiumum rank = {X_test["Rank"].max()}')
```
