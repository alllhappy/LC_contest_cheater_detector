convert into suitable data types and handling null values

```python

# df2=df results in same dataframe reference no copy is made
df2=df.copy()
df2['Q1TimeStamp'] = df2['Q1TimeStamp'].fillna(0).astype('int64')
df2['Q2TimeStamp'] = df2['Q2TimeStamp'].fillna(0).astype('int64')
df2['Q3TimeStamp'] = df2['Q3TimeStamp'].fillna(0).astype('int64')
df2['Q4TimeStamp'] = df2['Q4TimeStamp'].fillna(0).astype('int64')

df2['Q1FailCnt'] = df2['Q1FailCnt'].fillna(0).astype('int64')
df2['Q2FailCnt'] = df2['Q2FailCnt'].fillna(0).astype('int64')
df2['Q3FailCnt'] = df2['Q3FailCnt'].fillna(0).astype('int64')
df2['Q4FailCnt'] = df2['Q4FailCnt'].fillna(0).astype('int64')

df2['CurrRating'] = df2['CurrRating'].astype('int64')


df2.iloc[0:5]
```
